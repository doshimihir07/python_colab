```python
# prompt: Plot a graph between Secondary Education percentage and Higher Secondary Education percentage and categorize them based on the employment status. (hue = 'status'). add a best fit line

sns.lmplot(x='ssc_p', y='hsc_p', data=df, hue='status')
plt.show()
```

```python
df[df['status']=='Not Placed'].describe().T
```

```python
# prompt: 75% of not placed students have their Secondary Education percentage less than 65%. ?

not_placed_df = df[df['status'] == 'Not Placed']
percentage_less_than_65 = len(not_placed_df[not_placed_df['ssc_p'] < 65]) / len(not_placed_df) * 100
print(f"{percentage_less_than_65:.2f}% of not placed students have their Secondary Education percentage less than 65%")
```
